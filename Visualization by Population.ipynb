```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import median
```

```python
# Save file path to variable
population_income = "county_census_data.csv"
state_population = "county_census_debt_data_byState.csv"
debt_data = "county_census_debt_data.csv"
```

```python
debt_data_pd = pd.read_csv(debt_data)
debt_data_pd.head()
```

```python
debt_data_pd.plot(kind='bar',x='County',y='UnemploymentCount')
```

```python
x = debt_data_pd["Population"]
y = debt_data_pd["PovertyRate"]

plt.scatter(x,y)
plt.title("")
plt.xlabel("")
plt.ylabel("")


plt.savefig("Population_Income.png")
plt.show()
```

```python
state_population_pd = pd.read_csv(state_population)
state_population_pd.head()
```

```python
x = state_population_pd["PovertyRate"]
y = state_population_pd["StudentDebtAmount"]
plt.scatter(x,y)
plt.title("Student-Poverty")
plt.xlabel("Poverty Rate")
plt.ylabel("Student Debt Amount")


plt.savefig("Student-Poverty.png")
plt.show()
```

```python
x = state_population_pd["PovertyRate"]
y = state_population_pd["AutoDebtAmount"]
plt.scatter(x,y)
plt.title("Auto-Poverty")
plt.xlabel("Poverty Rate")
plt.ylabel("Auto Debt Amount")


plt.savefig("Auto-Poverty.png")
plt.show()
```

```python
x = state_population_pd["MedicalAmount"]
y = state_population_pd["HouseholdIncome"]
plt.scatter(x,y)
plt.title("Household-Medical")
plt.xlabel("Medical Amount Owed")
plt.ylabel("Household Income")


plt.savefig("Household-Medical.png")
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Medical-Poverty")
plt.xlabel("Poverty Rate")
plt.ylabel("Medical Amount")


plt.savefig("Medical-Poverty.png")
plt.show()
```
